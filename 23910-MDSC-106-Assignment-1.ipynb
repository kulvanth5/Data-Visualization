```python
from sklearn.datasets import load_diabetes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

#loading the diabetes dataset as a dataframe from sklearn
data = load_diabetes(as_frame=True)

#loading patient names and aadhar no. for the diabetes dataset from a real-world dataset
names = pd.read_excel('https://query.data.world/s/2no2ydgc53sprlii2eb4wk246pk3mj?dws=00000')
aadhar = pd.read_csv('https://global-uploads.webflow.com/62c841547dcbed63336d38a2/63bbd8f8cfc08416d58208b5_Aadhar-Number.csv')

#inserting name and aadhar no. into our dataset.
# Parameter True specifies to modify the existing dataframe rather than returning a new one.
data['data'].insert(0,'name of the patient',names['Name'],True)
data['data'].insert(1,'aadhar no.',aadhar[:442],True)

#copying the dataframe into another variable to use for plotting
df = data['data']

#this original is being used for learning pre-processing.
data['data']
```

inplace = true returns the modified version of existing dataframe rather than returning a new dataframe

plt.figure is used to the define the size of the total graph plotting.

plt.subplot is used to plot a single graph in the total plot.

1. Histplot plots the frequency of each value in a column and kde = true gives the density curve of the plot

2. scatter plot scatters the data-points wrt to x and y variables across the graph.

3. bar plot is used to plot a numerical value wrt to a categorical value.
In below plotting , I used age as the categorical and remaining features as numeric values. Since the data is normalized , age value ranges from -1 to 1.

```python
def plotting(df,v,t):

  # setting the total plot size
  plt.figure(figsize=(20,7))

  plt.subplot(1,3,1)
  sb.histplot(df[v],kde=True,color='r')
  plt.title('Histogram')

  plt.subplot(1,3,2)
  plt.scatter(df[v],df[t],color='g')
  plt.title('Scatter')

  plt.subplot(1,3,3)
  sb.barplot(x=df[v],y=df[t])
  plt.title('Bar Graph')


for feature in df:
  plotting(df,feature,'age')    #different types of plot for each feature
```

A box plot is a graphical representation of vastness and skewness of the data using quartiles. Typically , 50% the data lies in the inter-quartile range ,i.e., the box. And the line in the box is the median of the data.

The remaining 50% of the data is distributed eqaully on either side of the box and are known as quartile-1(Q1) and quartile-2(Q2).

```python
features = list(df.columns)

#removing below features from box plot
features.remove('sex')
features.remove('aadhar no.')

#categorizing sex value into 2 types. Say 1 for Male and 0 for Female
df.loc[df['sex'] > 0.0,'sex'] = 1
df.loc[df['sex'] < 0.0,'sex'] = 0

#box-plotting the range of values for each feature wrt to age ,i.e., male and female
def boxplot(df,x,y):

  plt.figure(figsize=(20,5))

  plt.subplot(1,3,1)
  sb.boxplot(x=df[y],y=df[x[0]],data=df)

  plt.subplot(1,3,2)
  sb.boxplot(x=df[y],y=df[x[1]],data=df)

  plt.subplot(1,3,3)
  sb.boxplot(x=df[y],y=df[x[2]],data=df)


while features:
  boxplot(df,features[:3],y='sex')  #passing 3 features at a time to above boxplot function to plot
  features = features[3:]
```

A stacked bar plot is used to compare values between two categories of data.

The main advantage of this is that we can compare the overall traits b/w the categories as well as across any particular features b/w the 2 categories.

For the diabetes dataset that I'm using , I divided the total no.of patient into two categories based on gender and the sex_count variable in below stores the count of no.of records wrt to each feature.

The avgs stores the average of all the values

The average value of each feature is plotted against the count of no.of patients based on their gender.

```python
#dropping aadhar no. from plotting.
stk = df.drop('aadhar no.',axis=1)

#filling the null values of the columns with the mean
stk['bmi'].fillna(value=stk['bmi'].mean(),inplace=True)
stk['s5'].fillna(value=stk['s5'].mean(),inplace=True)

def avg(df,param,f):
  return df.groupby(param)[f].mean()

#counting the no.of patients gender wise
sex_count = stk.groupby('sex').count()

#calculating average of various measurements.
avgs = [ avg(stk,'sex',f) for f in stk.columns if f != 'sex']

#to set bottom for each plot
csum = 0

#setting the size of the total plot and we plot each feature's mean value from avgs  with the gender-count.
plt.figure(figsize=(15,6))

for i in range(len(avgs)):

  #stacked bar plot of average of various features wrt gender count
  plt.subplot(1,2,1)
  plt.bar(sex_count.iloc[:,i],avgs[i],bottom=csum)      #iloc is used to access all the rows wrt a particular column.
  plt.legend(sex_count.columns)
  plt.title('Stacked Bar plot')
  csum = csum + avgs[i]

  plt.subplot(1,2,2)
  plt.bar(sex_count.iloc[:,i],avgs[i])              #normal bar plot of various features wrt to gender count
  plt.legend(sex_count.columns)
  plt.title('Bar plot')
```

From above plot , we can infer that

207 patients are male(assume that gender=1 is male) and their bmi ranges from 0.08 to 0.12 approximately. While 235 patient are female(assume gender==2 is female) and their bmi ranges from -0.06 to -0.1(normalised)
