```python
import pandas as pd

df1 = pd.read_csv('/content/TelcomCustomer-Churn_1.csv')
df2 = pd.read_csv('/content/TelcomCustomer-Churn_2.csv')

df = df1.merge(df2,on='customerID')

df.columns == df1.merge(df2).columns
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
df.info()
```

```python
import matplotlib.pyplot as plt

filter = ['customerID','MonthlyCharges','TotalCharges','tenure']
counts = []
i , j = 0 , 0

fig , ax = plt.subplots(5,4,figsize=(15,10))
fig.subplots_adjust(hspace=0.5)

for f in df.columns:

  if f not in filter:
    ax[i,j].pie(x=df.groupby(f).size(),labels=df[f].unique().tolist(),autopct='%1.1f%%',radius=1.2)
    ax[i,j].set_title(f)
    j = j + 1

  if j > 3:
    j = 0
    i = i + 1

```
