Below cell converts the columns which have only 0/1 as values to categorical types as per business definition

```python
for col in df.columns:

  # getting the unique values in each col of df
  values = df[col].unique()

  # checking if 0 & 1 are the only values present in them , if so convert them into object(categorical) type
  if list(values) == [0,1]:
    df[col] = df[col].astype('object')

# price_range and wifi are also categorical type and therefore converting it into object type
df['price_range'] = df['price_range'].astype('object')
df['wifi'] = df['wifi'].astype('object')
df.info()
```

**As per business definition , typically RAM is described in GB. So converting the RAM from 'x' MB's into 'n' GB's by dividing with 1024**

```python
df['ram'] = df['ram']/1024
```

**All the inferences are made by assuming the dataset covers all types of brands and are in accordance with the public demands**

```python
#visualizing % of different categories of mobile phones present in the df using pie-chart
sns.countplot(data=df,x='price_range')
```

From the above count plot , it can be seen that the dataset consists of all the mobile phones in equal numbers(500)

Therefore it can be inferred that the manufacturers cover all sections of consumers in terms of affordability

```python
# considering various categorical features in the mobiles for inferencing
cols=['four_g','touch_screen','dual_sim','three_g']

fig,axes=plt.subplots(2,2,figsize=(15,5))

for col,ax in zip(cols,axes.flatten()):

    labels = [f"{col}-supported",'Not supported']

    # calcuating the different values in the above columns
    values = df[col].value_counts().values

    # plotting the frequency of different values in pie chart for the above columns
    ax.pie(values, labels=labels, autopct='%1.1f%%',startangle=90,explode=[0.04,0.0])
    ax.set_title(f'{col}-Pie Chart')

# adjusting the spaces and gaps b/w the plots for good layout
plt.tight_layout()
```

**Pie charts are more useful in visualizing how much % of different values or types in a column or a feature are present by comparing them.**

**Hence , to see how many mobiles support/unsupport 4g, wifi, dual_sim , bluetooth, pie chart is used in the above case.**

Follwing inferences are made :-

52% of the phones support 4g. Nonetheless 48% of them are non 4g which means a decent no.of customers still use 3g.

50% of them are of touch_screen while 50% are not which means half of the customers use keypad mobiles which are really out of trend now.

51% of the customers use dual-sim mobiles which suggests that 51% of them have more than one mobile number

76% of the mobiles support 3g which means many people cannot afford or are good enough with 3g while 24% don't even use 3g which means they are a way behind in using the latest technology or in following the trends in technology.

```python
# joint plot (density plot) b/w RAM and price_range to see how price changes wrt RAM
sns.jointplot(x='ram',y='price_range',data=df,color='red',kind='kde',fill=True)
plt.title('joint plot for RAM vs price_range')
plt.tight_layout()
```

**To know whether the given dataset can be visualized as clusters , joint plot is very much handy and easier to do so comparative to any other plots and hence is used.**

 Therefore , as we can see above different price_range of mobiles can be divided into different clusters with respect to RAM(in GB).

From the above joint plot , it can be infered that RAM of

Category 3 > 2 > 1 > 0 in the order 4GB > 2.5GB > 2GB > 0.5GB respectively

and price of Category 3 > 2 > 1 > 0

That means as RAM increases in a mobile its price also increases

```python
# defining the figure size for sub plotting
fig , axes = plt.subplots(2,3,figsize=(15,10))
fig.subplots_adjust(hspace=0.3,wspace=0.3)

# considering a few continous features for detecting outliers in them
cols = ['int_memory','fc','mobile_wt','pc','sc_h','battery_power']

for col,ax in zip(cols,axes.flatten()):
  sns.boxplot(ax=ax,data=df,x='price_range',y=col,hue='price_range')
```

**To visualize outliers, distribution of the data and skewness, box plot is the most easiest way to do so against any other plots. And as we can see , the box plots also the various outliers wrt to different features in the mobile phones across different price ranges.**

plot(0,0): (int_memory vs price_range)
* Phones in the price range of 0, 1 and 2 offer similar ranges of internal memory. But phones in the price range of 3 offers a greater amount of internal memory as compared to the others which explains why those phones are costlier than the others.

* We can also see that there's no outliers when it comes to internal memory.

plot(0,1): (fc vs price_range)
* Most of the phones in the price range of 0 have a front camera of more than 2.5 pixels. There are few outliers which have a front camera of over 12.5 pixels. We can infer that even though the price of the phones are low, they offer a pretty decent camera quality.

* We can see that the boxplot of the phones in the price range 1, 2 and 3 have similar skewness and have pretty much identical distribution. Therefore, we can infer that the front camera on a phone in the price range of 1 and 2 can compete with the front camera of a phone in the price range of 3.

plot(0,2): (mobile_wt vs price_range)
* We can see that the weight distribution of the phones in the price range of 0,1 and 2 are somewhat similar. But the weight of the phones in the price range of 3 are little bit less. Therefore, we can infer that since many features are being incorporated in a phone of lesser weight compared to the others, the price are arguably higher.

* There are no outliers in any of the categories.

plot(1,0): (pc vs price_range)
* We can see that the primary camera range for phones in the price range of 0 and 1 are pretty much the same. But phones with primary camera in the price range of 2 and 3 offer higher qualities of primary cameras. Therefore, we can infer that the market has a higher demand for better qualities of primary camera as compared to front cameras.

* There are no outliers in any of the categories.

plot(1,1): (sc_h vs price_range)
* Phones in the price range of 0 and 3 have similar ranges of screen height but group 3 has a higher median than group 1.

* Similarly, phones in the price range of 1 and 2 have pretty similar ranges and median also.

* We can infer that the manufacturers are not focusing much on the screen size but rather on the performance of their phones.

plot(1,2): (battery_power vs price_range)
* We can see that phone in category 3 offers much higher battery power as compared to the other categories of phone. Therfore, we can deduce that battery power has a considerable impact on the price of a phone.

```python
# calculating the co-relation b/w different features in the dataset
co_relation = df.corr()

# plotting the co-relations using heatmap
plt.figure(figsize=(20,10))
sns.heatmap(co_relation,annot=True,linewidth=2)
plt.show()
```
